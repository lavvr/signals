```python
import numpy as np
import itertools

def build_h_matrix(m):
    cols = []
    for i in range(1, 2**m):
        bits = format(i, f'0{m}b')
        cols.append([int(b) for b in bits])
    return np.array(cols).T

def all_codewords(G):
    k = G.shape[0]
    messages = np.array(list(itertools.product([0, 1], repeat=k)))
    return np.dot(messages, G) % 2
```

```python
for m in [3, 4, 5]:
    n = 2**m - 1
    exp_weight = 2**(m - 1)

    print(f"\n{'='*40}")
    print(f"m = {m}  |  n = {n}  |  expected weight = {exp_weight}")
    print('='*40)

    H = build_h_matrix(m)
    cw = all_codewords(H)
    nz = cw[np.any(cw != 0, axis=1)]
    weights = np.sum(nz, axis=1)
    unique_weights = np.unique(weights)

    print("Unique weights of non‑zero codewords:", unique_weights)

    dists = set()
    for a, b in itertools.combinations(cw, 2):
        dists.add(np.sum(a != b))

    print("All pairwise distances:", dists)

    weight_ok = (len(unique_weights) == 1 and unique_weights[0] == exp_weight)
    dist_ok   = (len(dists) == 1 and exp_weight in dists)

    if weight_ok and dist_ok:
        print("The dual code is a simplex code.")
    else:
        print("Verification failed.")
```
